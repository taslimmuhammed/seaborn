```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"hr_data.csv")
df.head()
```

Dist plots are used for plotting histograms

```python
sns.distplot(df['average_montly_hours'])
```

Bins represents each interval in X axis

```python
bins = [2,3,4,5,6,7,8,10]
sns.distplot(df['time_spend_company'],bins=bins)
```

KDE- kernal Density Estimate [the Line we in between] \
hist - histogram\
rug = True :- more specefic visualisation

```python
sns.distplot(df['time_spend_company'],bins=bins, kde=False,rug=True)
```

# KWS
hist_kws for customizing histogram\
kde_kws for customizing line\
params:-\
color, edgecolor, linewidth, alpha[transparency]


```python
sns.distplot(df['time_spend_company'],bins=bins,rug=True,
            hist_kws={'color':'red', 'edgecolor':'blue', 'linewidth':3, 'alpha':0.5},
            kde_kws={'color':'red', 'linewidth':3}
            )
```

```python
sns.distplot(df['time_spend_company'],bins=bins,rug=True,
            color='green',vertical=True
            )
```
