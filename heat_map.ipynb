```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df = df.pivot(index='month',columns='year',values='passengers')
df
```

```python
sns.heatmap(df)
```

Annot = Annotation\
fmt = format [decimals here]

```python
plt.figure(figsize=(14,6))
sns.heatmap(df, annot=True, fmt='d')
```

# custom lines


```python
plt.figure(figsize=(14,6))
sns.heatmap(df, fmt='d', annot=True, linewidth='5', linecolor='k', cmap='Blues')
```

cbar = lines in between [true/false]

```python
sns.heatmap(df,cbar=False)
```

```python
grid_kws = {'height_ratios':(.3,.05), 'hspace':.4}
f,(ax,cbar_ax) = plt.subplots(2,gridspec_kw = grid_kws)
ax = sns.heatmap(df,cbar_kws={"orientation":"horizontal"},ax = ax, cbar_ax = cbar_ax)
```
