```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=df, hue='class',style='alive', palette='gist_rainbow', alpha=0.5)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='fare', data=df, hue='class',style='alive', palette='gist_rainbow')
sns.lineplot(x='age', y='fare', data=df, color='green')
```
